```python
# Since the Traget column "SalePrice" is right skewed. We Can Handle it.
df["log_SalePrice"]=np.log1p(df["SalePrice"])
# As Outcome we will see improvement in the RMSE and R2, and Makes the errors more balanced and  reduces skewness
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['log_SalePrice'], kde=True)
plt.show()
```

```python
# Previously we got the right skewed distribution on this column and now it is normmally distributed.
```

```python
# High Missing
high_missing=missing[missing>60]
high_missing
# These are the four columns with more than 60% of missing values.
```

```python
# Except the MiscFeature. we handle the null values by replacing none, because the other column are a bit important than this.
```

```python
for col in ["PoolQC","Alley","Fence"]:
    df[col]=df[col].fillna("None")
```

```python
# We can drop the MiscFeature Column
df.drop(columns="MiscFeature",axis=1,inplace=True)
```

```python
df.isna().sum().sort_values(ascending=False).head(15)
```

```python
# Similarly we can handle the null value in the remaining columns also.
cat_none_cols = [
    'PoolQC', 'Alley', 'Fence', 'FireplaceQu',
    'GarageQual', 'GarageFinish', 'GarageType', 'GarageCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
    'BsmtCond', 'BsmtQual',
    'MasVnrType'] # These are the columns where we need to fill the null values with "None".
for col in cat_none_cols:
    df[col]=df[col].fillna("None")
```

```python
num_zero_cols = [
    'MasVnrArea',
    'GarageYrBlt'] # These are the columns where we need to fill the null values with 0.

for col in num_zero_cols:
    df[col] = df[col].fillna(0)
```

```python
# "LotSize" varies by neighbour,So, Global median for "LotFrontage" would distort local patterns
df["LotFrontage"] = df.groupby('Neighborhood')["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

```python
# this column had to be replaced with the mode value
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
# Now we have handled all the null values in the Data. 
df.isna().sum().sort_values(ascending=False).head()
```

```python
# Encoding the categorical columns
df=pd.get_dummies(df,drop_first=True)
```

```python
df.shape
```

```python
x=df.drop(columns=["SalePrice","log_SalePrice"])
y=df["log_SalePrice"]
```

```python
# Now let's start to Build the model
from sklearn.model_selection import train_test_split # Used to split and train a model

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=42)
```

```python
# Required Libraries
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.001),
    "Random Forest": RandomForestRegressor(n_estimators=200,random_state=42,n_jobs=-1) # n_jobs is used to utilize the CPU cores
}
```

```python
for model_name, model in models.items():
    
    # Train
    model.fit(x_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    
    # Metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    # Output
    print("="*40)
    print(f"Model: {model_name}")
    print(f"Train RMSE: {train_rmse:.4f}")
    print(f"Test  RMSE: {test_rmse:.4f}")
    print(f"Train R2  : {train_r2:.4f}")
    print(f"Test  R2  : {test_r2:.4f}")
```

```python
# Linear Regression and RandomForest has got overfitted
# Lasso doesn't perform well as Ridge
# Ridge Regression was selected as the final model because it achieved the lowest test RMSE and highest test RÂ² 
# while maintaining a balanced gap between training and testing performance, indicating strong generalization on unseen data.
```

```python
# Since we Finalized that the Ridge model is the best we can try to perform Tuning and let's see is there any improvement or not.
from sklearn.model_selection import GridSearchCV
ridge_params = {"alpha": [0.01, 0.1, 1, 5, 10, 20, 50, 100]}
ridge_tune=Ridge()
```

```python
ridge_grid = GridSearchCV(
    estimator=ridge_tune,
    param_grid=ridge_params,
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)

ridge_grid.fit(x_train, y_train)
```

```python
ridge_grid.best_params_
ridge_grid.best_score_
```

```python
best_ridge = ridge_grid.best_estimator_

y_train_pred = best_ridge.predict(x_train)
y_test_pred = best_ridge.predict(x_test)

train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

print(f"train_rmse : {train_rmse}\ntest_rmse :{test_rmse}\ntrain_r2 :{train_r2}\ntest_r2 :{test_r2}")
```

```python
# finally we found that the tuning has no improvement and the test data performance has droped a bit.
# So we can keep the first model of Ridge for our Project.
```

```python
# Final Model then above models are for Comparing.
final_model = Ridge(alpha=1.0)
final_model.fit(x, y)
```
